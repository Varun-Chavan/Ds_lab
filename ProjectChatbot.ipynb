```python
# Function to parse the chatbot.txt file
def load_chatbot_data(file_path):
    qa_dict = {}
    with open(file_path, 'r') as file:
        lines = file.readlines()
        question = ""
        answer = ""
        for line in lines:
            line = line.strip()
            if line and line[0].isdigit():
                if question and answer:
                    qa_dict[question.lower()] = answer
                question = line.split('. ', 1)[1].strip()
                answer = ""
            elif line.startswith('-'):
                answer = line[1:].strip()
        if question and answer:
            qa_dict[question.lower()] = answer
    return qa_dict

# Load the data from the file
file_path = "/content/chatbot.txt"  # Replace with the correct path if different
qa_dict = load_chatbot_data(file_path)

# Print the loaded Q&A pairs to verify
for q, a in qa_dict.items():
    print(f"Q: {q}")
    print(f"A: {a}\n")
```

```python
from difflib import get_close_matches

def get_response(user_input, qa_dict):
    user_input = user_input.lower()
    questions = list(qa_dict.keys())
    closest_match = get_close_matches(user_input, questions, n=1, cutoff=0.5)
    if closest_match:
        return qa_dict[closest_match[0]]
    return "I'm sorry, I don't have an answer for that."
```

```python
def get_response(user_input, qa_dict):
    user_input = user_input.lower()
    questions = list(qa_dict.keys())
    print(f"Available Questions: {questions}")  # Debug: Print available questions
    closest_match = get_close_matches(user_input, questions, n=1, cutoff=0.5)
    print(f"User Input: {user_input}")  # Debug: Print user input
    print(f"Closest Match: {closest_match}")  # Debug: Print closest match
    if closest_match:
        return qa_dict[closest_match[0]]
    return "I'm sorry, I don't have an answer for that."
```

```python
# Example usage
user_inputs = [
    "What services do you offer?",
    "How can I contact support?",
    "What are your business hours?",
    "Tell me about your services.",
    "Can I get support?",
    "What time are you open?"
]

for user_input in user_inputs:
    response = get_response(user_input, qa_dict)
    print(f"User Input: {user_input}")
    print(f"Response: {response}\n")
```

```python
import gradio as gr  # Ensure you import gradio

# Define the Gradio interface
def chatbot_interface(user_input):
    return get_response(user_input, qa_dict)

# Create the Gradio interface
iface = gr.Interface(fn=chatbot_interface,
                     inputs=gr.Textbox(label="Ask me anything", placeholder="Enter your question here..."),
                     outputs=gr.Textbox(label="Response"),
                     title="How can I help you?",
                     description="Ask questions and get answers based on the data in the chatbot.txt file.")

# Launch the Gradio app
iface.launch(share=True)

```
